### HBase tables:
    CREATE 'tweets', 'tweet'
    CREATE 'neg_counter', 'info'
    CREATE 'pos_counter', 'info'

### Hive external tables:
    CREATE EXTERNAL TABLE tweets(key string, text string, cleaned_text string, source string, lang string, target string) 
    STORED BY 'org.apache.hadoop.hive.hbase.HBaseStorageHandler' 
    WITH SERDEPROPERTIES ("hbase.columns.mapping" = ":key,tweet:text,tweet:cleaned_text,tweet:source,tweet:lang,tweet:target")  
    TBLPROPERTIES ("hbase.table.name" = "tweets");

    CREATE EXTERNAL TABLE neg_counter(key string, count bigint) 
    STORED BY 'org.apache.hadoop.hive.hbase.HBaseStorageHandler' 
    WITH SERDEPROPERTIES ("hbase.columns.mapping" = ":key,info:counter#b") 
    TBLPROPERTIES ("hbase.table.name" = "neg_counter");

    CREATE EXTERNAL TABLE pos_counter(key string, count bigint) 
    STORED BY 'org.apache.hadoop.hive.hbase.HBaseStorageHandler' 
    WITH SERDEPROPERTIES ("hbase.columns.mapping" = ":key,info:counter#b") 
    TBLPROPERTIES ("hbase.table.name" = "pos_counter");

```python
import os  
os.environ['PYSPARK_SUBMIT_ARGS'] = '--jars /usr/lib/hive/lib/hive-hbase-handler.jar,/usr/lib/hbase/lib/*.jar pyspark-shell'  

from os.path import expanduser, join, abspath
from pyspark import SparkContext
from pyspark.sql import SparkSession
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

spark = SparkSession.builder.appName("Sentiment Query") \
    .config("hive.metastore.uris", "thrift://localhost:9083") \
    .enableHiveSupport().getOrCreate()
```

```python
query = spark.sql("SELECT key, counter FROM pos_counter WHERE LENGTH(key) >= 3  ORDER BY counter DESC LIMIT 20")
```

```python
df = query.toPandas()
df.plot(kind='barh', x='key', title="Top 20 positive words", legend=False)
```

```python
query = spark.sql("SELECT key, counter FROM neg_counter WHERE LENGTH(key) >= 3 ORDER BY counter DESC LIMIT 20")
```

```python
df = query.toPandas()
df.plot(kind='barh', x='key', title="Top 20 negative words", legend=False)
```

```python
query = spark.sql("SELECT target, count(target) AS count FROM tweets GROUP BY target")
df = query.toPandas()
df.plot(kind='pie', labels=df['target'], y='count', autopct='%1.1f%%', startangle=90, legend=False)
```

### Word Cloud: Positive words

```python
query = spark.sql("SELECT key, counter FROM pos_counter ORDER BY counter DESC LIMIT 100")
df = query.toPandas()

freqs = dict(zip(df['key'].values, df['counter'].values))

wordcloud = WordCloud(width=1600, height=800,max_font_size=200).generate_from_frequencies(freqs)
plt.figure(figsize=(12,10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

### Word cloud: Negative words

```python
query = spark.sql("SELECT key, counter FROM neg_counter ORDER BY counter DESC LIMIT 100")
df = query.toPandas()

freqs = dict(zip(df['key'].values, df['counter'].values))
wordcloud = WordCloud(width=1600, height=800,max_font_size=200).generate_from_frequencies(freqs)
plt.figure(figsize=(12,10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
